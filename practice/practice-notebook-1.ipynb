15. Find factorial using loop.

```python
def factorial_for_loop(n):
    if n < 0:
        return "factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

num = 5
print(f"Factorial of a number: {factorial_for_loop(num)}")
```

```python
def route_query(query : str) -> str:
    q = query.lower()
    if "cancel" in q or "refund" in q : return "human_agent"
    elif "track" in q or "where" in q : return "order_agent"
    elif "return" in q or "exchange" in q : return "returns_agent"
    elif "price" in q or "discounts" in q : return "promotions_agent"
    elif "product" in q or "spec" in q : return "catalog_agent"
    else:
        return "general_agent"
queries = [
    "Where is my order?",
    "I want a refund",
    "Do you have discounts?"
]

for q in queries:
    print(f" '{q}'  -> {route_query(q)}")
```
