## 1 — The Problem Without Classes

Imagine you are storing data about a product review. Without a class you end up with loose variables:


````python
# Without a class — loose variables for ONE review
review_id   = 5001
rating      = 4
title       = 'Really good monitor'
verified    = True

# A function that works on those variables
def get_stars(rating):
    return '★' * rating + '☆' * (5 - rating)

print(get_stars(rating), title)
````

````python
# Now you have a SECOND review — you need a whole new set of variables
review_id_2   = 5002
rating_2      = 2
title_2       = 'Disappointed'
verified_2    = False

print(get_stars(rating_2), title_2)

# Problems:
# 1. Variables are not grouped — easy to mix up rating and rating_2
# 2. The function get_stars() is floating — not connected to the data
# 3. For 1000 reviews this becomes completely unmanageable
````

**The solution:** bundle the data AND the functions that work on it together in one place — a **class**.

## 2 — Your First Class

A class is defined with the `class` keyword. Think of it as a **blueprint** — you define it once, then create as many objects (instances) from it as you need.

```python
class ClassName:
    def method(self):
        ...
```

| Python | Java / C# equivalent |
|--------|---------------------|
| `class MyClass:` | `public class MyClass { }` |
| `def __init__(self)` | constructor |
| `self` | `this` |
| `obj = MyClass()` | `obj = new MyClass()` |


````python
# The simplest possible class — just a container
class ProductReview:
    pass   # pass = empty body — valid Python

# Create an object (instance) from the class
# In Java/C# you would write: ProductReview r = new ProductReview()
# In Python — no 'new' keyword:
review = ProductReview()

print(type(review))          # <class '__main__.ProductReview'>
print(isinstance(review, ProductReview))  # True
````

````python
# You can attach data to an object using dot notation
review.review_id = 5001
review.rating    = 4
review.title     = 'Really good monitor'

print(review.rating)   # 4
print(review.title)    # Really good monitor
````

> Attaching attributes this way works but is messy — you have to remember to set every field manually every time.  
> That is what `__init__` is for.

## 3 — `__init__` — The Constructor

`__init__` is called **automatically** when you create an object. It sets up all the starting data.

- In **Java/C#**: `__init__` is the constructor method
- **`self`** is the object being created — it refers to *this specific instance*
- `self.name = value` creates an **instance variable** — each object gets its own copy

```python
class MyClass:
    def __init__(self, value):   # self = the new object being built
        self.value = value       # self.value = instance variable
```


````python
class ProductReview:

    def __init__(self, review_id, rating, title, verified=False):
        """
        __init__ is called automatically when you write ProductReview(...).
        self = the new object being created.
        self.rating = instance variable — each review gets its own rating.
        """
        self.review_id = review_id
        self.rating    = rating
        self.title     = title.strip()   # clean whitespace on the way in
        self.verified  = verified        # False by default


# Create two reviews — each has its own independent data
# __init__ is called automatically here:
review1 = ProductReview(5001, 4, 'Really good monitor')
review2 = ProductReview(5002, 2, 'Disappointed', verified=True)

print(review1.rating, review1.title)
print(review2.rating, review2.title, review2.verified)
````

### What `self` actually is

`self` is just the object itself — Python passes it automatically as the first argument to every method.

```python
review1 = ProductReview(5001, 4, 'Great')
#                  ↑
# Python creates a new object and passes it as 'self'
# Inside __init__: self IS review1
#   self.rating = 4   →   review1.rating = 4
```

> **Java/C# note:** In Python you must always write `self` explicitly as the first parameter.  
> In Java you have `this` available implicitly — in Python `self` must be declared.

## 4 — Instance Methods

Methods are **functions defined inside a class**. They always receive `self` as their first parameter — this gives them access to the object's data.

```python
class MyClass:
    def my_method(self):   # self gives access to instance variables
        return self.value
```


````python
class ProductReview:

    def __init__(self, review_id, rating, title, verified=False):
        self.review_id = review_id
        self.rating    = rating
        self.title     = title.strip()
        self.verified  = verified

    # Instance method — has access to self (the object's data)
    def get_stars(self):
        """Return the rating as filled/empty stars. e.g. ★★★☆☆"""
        return '★' * self.rating + '☆' * (5 - self.rating)

    def is_positive(self):
        """Return True if rating is 4 or above."""
        return self.rating >= 4

    def to_context(self):
        """Format this review as a context block for the LLM prompt."""
        verified_tag = '[Verified] ' if self.verified else ''
        return (
            f'<review>\n'
            f'  Rating  : {self.get_stars()} ({self.rating}/5)\n'
            f'  Title   : {verified_tag}{self.title}\n'
            f'</review>'
        )


# Create a review
r = ProductReview(5001, 4, '  Really good monitor  ', verified=True)

# Call its methods — notice: no argument needed, self is passed automatically
print(r.get_stars())     # ★★★★☆
print(r.is_positive())   # True
print()
print(r.to_context())
````

### How method calls work under the hood

```python
r.get_stars()          # what you write
ProductReview.get_stars(r)   # what Python actually does — passes r as self
```

This is why `self` is always the first parameter — Python automatically fills it in with the object you called the method on.


````python
# Both are identical — Python translates the first into the second
print(r.get_stars())
print(ProductReview.get_stars(r))   # same result
````

## 5 — `__repr__` — How an Object Displays Itself

Without `__repr__`, printing an object gives you a useless memory address:
```
<__main__.ProductReview object at 0x7f3a4b2c1d50>
```

`__repr__` lets you define what the object looks like when printed or logged.

- In **Java**: `toString()` method
- In **C#**: `ToString()` method
- In **Python**: `__repr__`


````python
class ProductReview:

    def __init__(self, review_id, rating, title, verified=False):
        self.review_id = review_id
        self.rating    = rating
        self.title     = title.strip()
        self.verified  = verified

    def get_stars(self):
        return '★' * self.rating + '☆' * (5 - self.rating)

    def is_positive(self):
        return self.rating >= 4

    def to_context(self):
        verified_tag = '[Verified] ' if self.verified else ''
        return (
            f'<review>\n'
            f'  Rating  : {self.get_stars()} ({self.rating}/5)\n'
            f'  Title   : {verified_tag}{self.title}\n'
            f'</review>'
        )

    def __repr__(self):
        """
        Called when you print(review) or log the object.
        Java/C# equivalent: toString()
        Convention: show the key fields that identify this object.
        """
        return (
            f'ProductReview('
            f'id={self.review_id}, '
            f'rating={self.rating}, '
            f'verified={self.verified})'
        )


r1 = ProductReview(5001, 4, 'Really good monitor', verified=True)
r2 = ProductReview(5002, 2, 'Disappointed')

print(r1)          # uses __repr__
print(r2)
print([r1, r2])    # __repr__ used inside a list too
````

## 6 — `@property` — Access a Method Like a Variable

`@property` turns a method into an attribute. The caller reads it **without parentheses** — it looks like a variable but runs code.

```python
# Without @property — caller must use ():
r.get_stars()

# With @property — reads like a variable:
r.stars
```

Use `@property` when:
- The value is **computed** from other instance variables
- The name reads more naturally as a noun than as a verb
- The computation is cheap

| Python | Java / C# |
|--------|----------|
| `@property` | getter method / computed property |
| `r.stars` | `r.getStars()` or `r.Stars` |


````python
class ProductReview:

    def __init__(self, review_id, rating, title, verified=False):
        self.review_id = review_id
        self.rating    = rating
        self.title     = title.strip()
        self.verified  = verified

    @property
    def stars(self):
        """Rating as filled/empty stars. Access as r.stars (no parentheses)."""
        return '★' * self.rating + '☆' * (5 - self.rating)

    @property
    def is_positive(self):
        """True if rating is 4 or above."""
        return self.rating >= 4

    @property
    def summary(self):
        """One-line summary — useful for logging and prompt injection."""
        badge = '[Verified] ' if self.verified else ''
        return f'{self.stars} {badge}{self.title}'

    def to_context(self):
        return (
            f'<review>\n'
            f'  Rating  : {self.stars} ({self.rating}/5)\n'
            f'  Title   : {self.title}\n'
            f'</review>'
        )

    def __repr__(self):
        return f'ProductReview(id={self.review_id}, rating={self.rating}, verified={self.verified})'


r = ProductReview(5001, 4, 'Really good monitor', verified=True)

# @property — access without parentheses
print(r.stars)        # ★★★★☆
print(r.is_positive)  # True
print(r.summary)      # ★★★★☆ [Verified] Really good monitor

# Regular method — needs parentheses
print(r.to_context())
````

````python
# @property also protects against accidental overwrite
r2 = ProductReview(5002, 2, 'Disappointed')

# r2.stars = '★★★★★'   # AttributeError — can't set a property without a setter
# This is intentional — stars is computed from rating, not stored separately

# Changing the source data updates the property automatically
print('Before:', r2.stars)
r2.rating = 5
print('After :', r2.stars)   # recomputed from new rating
````

## 7 — Working with a List of Objects

In production, you load many reviews from CSV and create a list of objects.  
The same patterns from Day 02 (loops, list comprehensions) apply to lists of objects.


````python
# Build a list of ProductReview objects from raw data
# (In production this comes from csv.DictReader)
raw_reviews = [
    {'review_id': 5001, 'rating': 5, 'title': 'Excellent!',     'verified': True},
    {'review_id': 5002, 'rating': 2, 'title': 'Disappointed',   'verified': False},
    {'review_id': 5003, 'rating': 4, 'title': 'Really good',    'verified': True},
    {'review_id': 5004, 'rating': 1, 'title': 'Terrible',       'verified': False},
    {'review_id': 5005, 'rating': 5, 'title': 'Perfect product','verified': True},
]

# Create a list of objects — one per raw dict
# ITERATION TRACE:
# Iter 1 → r={'review_id':5001,'rating':5,'title':'Excellent!','verified':True}
#           ProductReview(5001, 5, 'Excellent!', True) → object appended
# Iter 2 → r={'review_id':5002,'rating':2,...} → ProductReview(5002, 2, ...) appended
# ...
reviews = [
    ProductReview(r['review_id'], r['rating'], r['title'], r['verified'])
    for r in raw_reviews
]

# Print all reviews using __repr__
for rev in reviews:
    print(rev)
````

````python
# Filter — only positive reviews (rating >= 4)
# ITERATION TRACE:
# review_id=5001  is_positive = 5>=4 = True  → include
# review_id=5002  is_positive = 2>=4 = False → skip
# review_id=5003  is_positive = 4>=4 = True  → include
# review_id=5004  is_positive = 1>=4 = False → skip
# review_id=5005  is_positive = 5>=4 = True  → include
positive = [r for r in reviews if r.is_positive]

print(f'Positive reviews: {len(positive)} out of {len(reviews)}')
for r in positive:
    print(f'  {r.summary}')
````

````python
# Sort by rating (highest first)
# sorted() uses the rating @property via the key= lambda
by_rating = sorted(reviews, key=lambda r: r.rating, reverse=True)

print('Reviews by rating (highest first):')
for r in by_rating:
    print(f'  {r.stars}  {r.title}')
````

````python
# Build a context block for the LLM prompt from all positive verified reviews
good_and_verified = [
    r for r in reviews
    if r.is_positive and r.verified
]

context_block = '\n'.join(r.to_context() for r in good_and_verified)
print(context_block)
````

## Summary — Day 06

| Concept | Python | Java / C# equivalent |
|---------|--------|---------------------|
| Define a class | `class MyClass:` | `public class MyClass { }` |
| Constructor | `def __init__(self, ...)` | Constructor method |
| `self` | First param of every method | `this` |
| Create object | `obj = MyClass(...)` | `obj = new MyClass(...)` |
| Instance variable | `self.name = value` | `this.name = value` |
| Instance method | `def method(self):` | Regular method |
| `__repr__` | `def __repr__(self):` | `toString()` / `ToString()` |
| `@property` | `@property` on a method | getter / computed property |

**Key rules:**
- `self` must always be the first parameter of every method — Python passes the object automatically
- `__init__` runs automatically when you create an object — use it to set up instance variables
- Each object gets its own copy of instance variables — changing `review1.rating` does not affect `review2.rating`
- `@property` lets callers read `r.stars` without `()` — reads like a variable, runs code internally
- `__repr__` defines what `print(obj)` shows — make it useful for debugging
